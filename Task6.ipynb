```python
import pandas as pd
     

ds = pd.read_csv("https://raw.githubusercontent.com/insignificantGuy/Kaggle-Datasets/main/csgo_round_snapshots.csv")
     

ds.head(15)
```

```python
ds.info()
```

```python
ds.isna().sum()
```

```python
data = pd.get_dummies(data=ds,drop_first=True)
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
array = scaler.fit_transform(data.drop('round_winner_T',axis=1))
dif = pd.DataFrame(array,columns=data.columns[:-1])
dif
```

```python
x = dif
y = data['round_winner_T']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state = 100)
from sklearn.neighbors import KNeighborsClassifier
knc = KNeighborsClassifier()
knc.fit(x_train,y_train)
```

```python
predict = knc.predict(x_test)
from sklearn.metrics import classification_report, confusion_matrix
matrix = confusion_matrix(y_test,predict)
matrix
```

```python
print(classification_report(y_test,predict))
```
